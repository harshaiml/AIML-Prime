# Relational Database: SQL

## <u>What is a Database?</u>
It is a collection of data in a format that can be easily accessed.

Why databases?
- Can store large data
- Features like security, scalability etc.
- Easier to insert, update or delete data

## <u>SQL v/s NoSQL</u>

1. SQL

- Relational Database (data stored in Tables)

- eg - MySQL, Oracle, PostgreSQL etc.

2. NoSQL

- Non Relation Database (data stored in document/key-vall/graphs etc.)

- eg - MongoDb, Cassandra, PostgreSQL etc.

SQL is a programming language used to intrect with relational databases.

```python
sql = """
INSERT INTO user
(id, age, name, email, followers, following)
VALUES
(%s, %s, %s, %s, %s, %s)
"""

values = [
    (1, 14, "Adam", "adam@yahoo.in", 123, 145),
    (2, 15, "Bob", "bob123@gmail.com", 200, 200),
    (3, 16, "Casey", "casey@gmail.com", 300, 306),
    (4, 17, "Donald", "donald@gmail.com", 200, 105)
]

mycursor.executemany(sql, values)

mydb.commit()

print("Data Inserted Successfully!")
```

```python
mycursor.execute("""
CREATE TABLE Student (
    rollno INT PRIMARY KEY,
    name VARCHAR(30),
    city VARCHAR(30),
    mark INT,
    grade VARCHAR(5)
)
""")

print("Student Table Created")
```

```python
mycursor.execute("""
INSERT INTO Student
(rollno, name, city, mark)
VALUES
(110, "Adam", "Delhi", 76),
(108, "Bob", "Mumbai", 65),
(124, "Casey", "Pune", 94),
(112, "Duke", "Pune", 80)
""")

mydb.commit()

print("Student Data Inserted")
```

```python
mycursor.execute("""
SELECT DISTINCT city
FROM Student
""")

result = mycursor.fetchall()

for row in result:
    print(row)
```

```python
mycursor.execute("""
SELECT city, MAX(mark)
FROM Student
GROUP BY city
""")

result = mycursor.fetchall()

for row in result:
    print(row)
```

```python
mycursor.execute("""
SELECT AVG(mark)
FROM Student
""")

result = mycursor.fetchall()

print(result)
```

```python
mycursor.execute("SET SQL_SAFE_UPDATES = 0")

print("Safe Updates Disabled")
```

```python
mycursor.execute("""
UPDATE Student
SET grade =
CASE
    WHEN mark >= 80 THEN "O"
    WHEN mark >= 70 THEN "A"
    WHEN mark >= 60 THEN "B"
END
""")

mydb.commit()

print("Grades Updated")
```
