```python
import pandas as pd

# Load data
df = pd.read_csv("products.csv")

print("Original Data:")
print(df)

# -----------------------------
# 1. Basic Info
# -----------------------------
print("\nData Info:")
print(df.info())

# -----------------------------
# 2. Missing Values
# -----------------------------
print("\nMissing Values:")
print(df.isnull().sum())

# Fill missing rating with average
df['rating'] = df['rating'].fillna(df['rating'].mean())

# -----------------------------
# 3. Remove Duplicates
# -----------------------------
df = df.drop_duplicates()

# -----------------------------
# 4. Basic Analysis
# -----------------------------
print("\nAverage Price:", df['price'].mean())
print("Average Rating:", df['rating'].mean())

# -----------------------------
# 5. Filter Data
# -----------------------------
high_rated = df[df['rating'] > 4]
print("\nHigh Rated Products:")
print(high_rated)

# -----------------------------
# 6. GroupBy (Important 🔥)
# -----------------------------
category_analysis = df.groupby('category')['rating'].mean()
print("\nAverage Rating by Category:")
print(category_analysis)

# -----------------------------
# 7. Sorting
# -----------------------------
top_products = df.sort_values(by='rating', ascending=False)
print("\nTop Products:")
print(top_products.head())

# -----------------------------
# 8. Save Cleaned Data
# -----------------------------
df.to_csv("cleaned_products.csv", index=False)

print("\nCleaned data saved!")
```
