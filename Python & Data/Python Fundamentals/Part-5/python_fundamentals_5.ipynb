## <u>File Operations</u>
- r ------------> reading [default]
- w ------------> writing, truncates file first
- x ------------> creates new & open for writing
- a ------------> writing. appends at end
- b ------------> binary mode
- t ------------> text mode [default]
- #+------------> opens disk file for update (r & w)

## <u>Word Search</u>

```python
data = True
line = 1
word = input("Enter your value:")

with open("sample.txt", "r") as f:
    while data:
        data = f.readline()

        if("Python" in data):
            print(f"{word} word found at line {line}")
            break

        print(data)
        line += 1
```

## <u>Exception Handling</u>
try, except, else, finally

```python
try:
    x = int(input("Enter x: "))
    ans = 10/x

except ZeroDivisionError:
    print(f"Divide by 0 is not allowed")

else:
    print(f"Answer = {ans}")
```

```python
try:
    x = int(input("Enter x: "))
    ans = 10/x

except ZeroDivisionError:
    print(f"Divide by 0 is not allowed")

except ValueError:
    print(f"Invalid input")

else:
    print(f"Answer = {ans}")
```

```python
# finally keyword

try:
    x = int(input("Enter x: "))
    ans = 10/x

except ZeroDivisionError:
    print(f"Divide by 0 is not allowed")

except ValueError:
    print(f"Invalid input")

else:
    print(f"Answer = {ans}")

finally:
    print("End of program")
```

## <u>List Comprehensions</u>
[output for item in iterable if condition]

```python
squares = []

for i in range(6):
    squares.append(i*i)

print(squares)
```

```python
squares = []

for i in range(6):
    squares.append(i*i)

print(squares)

sq = [i*i for i in range(6)]
print(sq)
```

```python
# print odd numbers square

sq = [i*i for i in range(6) if i%2 != 0]
print(sq)
```

```python
nums = [-2, -3, 3, 4, -1, 7]

nums = [0 if val < 0 else val for val in nums]
print(nums)
```

```python
words = ["hello", "pyhton",]
print(words[0].upper())
```

```python
words = ["hello", "pyhton",]

words = [val.upper() for val in words]
print(words)
```

## <u>JSON Module</u>
- json.loads()
- json.dumps()
- json.load()
- json.dump()

```python
import json

json_str = '{"name": "Harsh", "isTeacher": true}'

py_obj = json.loads(json_str)

print(type(py_obj), py_obj)
```

```python
json_str = '{"name": "Harsh", "isTeacher": null}'

py_obj = json.loads(json_str)

print(type(py_obj), py_obj)
```

```python
py_obj = {
    "name": "Harsh",
    "isTeacher": True
}

json_str = json.dumps(py_obj)

print(type(json_str), json_str)
```

```python
py_obj = {
    "name": "Harsh",
    "isTeacher": None
}

json_str = json.dumps(py_obj)

print(type(json_str), json_str)
```

```python
with open("data.json", "r") as f:
    py_obj = json.load(f)
    print(py_obj)
```

```python
with open("data.json", "r") as f:
    py_obj = json.load(f)
    print(type(py_obj))
```

```python
data = {
    "name": "Shradha",
    "age": 24,
    "isTeacher": True
}

with open("data.json", "w") as f:
    json.dump(data, f)
```

```python
data = {
    "name": "Shradha",
    "age": 24,
    "isTeacher": True
}

with open("data.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
data = {
    "name": "Shradha",
    "age": 24,
    "isTeacher": True
}

with open("data.json", "w") as f:
    json.dump(data, f, indent=4, sort_keys=True)
```
