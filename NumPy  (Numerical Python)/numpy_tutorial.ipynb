
### Python List vs NumPy Arrays

1. NumPy arrays are faster (optimized in C)
2. NumPy stores data in continuous memory blocks
3. Uses less memory (optimized storage)
4. Built-in mathemetical functions
5. Supports dicrect vectorized operation

```python
# Execution performance
size = 100_000_000

# Python list
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")

# NumPy arrays
np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2
end = time.time()
print(f"numpy array time = {end-start} seconds")
```

```python
# Memory
import sys

print(f"python list size = {sys.getsizeof(py_list) * len(py_list)} bytes")

print(f"numpy array size = {np_arr.nbytes} bytes")
```

### So why is NumPy faster?
- Homogeneous Data Types and Contiguous Memory
- Vector Operations & C Implementations 
- Redued Overhead

### Operations on NumPy Arrays
- Reshaping
- Indexing (1D & 2D array)
- Fancy & Boolean Indexing
- Slicing

```python
# Operations on arrays

arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr, arr.shape)

reshaped = arr.reshape((3, 2))
print(reshaped, reshaped.shape)

flattend = arr.flatten() # 2D => 1D
print(flattend, flattend.shape)
```

```python
# Indexing
arr = np.array([1, 2, 3, 4, 5])

print(arr[0])
print(arr[3])
```

```python
# Indexing
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr[1][1])
```

```python
# Fancy Indexing
arr = np.array([1, 2, 3, 4, 5])

idx = [0, 1, 3]
print(arr[idx])
```

```python
# Boolean Indexing
arr = np.array([1, 2, 3, 4, 5])

print(arr[arr > 2])
print(arr[arr % 2 == 0])
print(arr[arr % 2 != 0])
```

```python
# Slicing
arr = np.array([1, 2, 3, 4, 5])

print(arr[1:4]) #[start, end, step]
print(arr[1:])
print(arr[:4])
print(arr[::2])
```

```python
# Copy vs View
nums = [1, 2, 3, 4, 5]
sub_list = nums[1:3]
print(sub_list)
sub_list[0] = 200

print(sub_list)
print(nums)
```

```python
arr = np.array([1, 2, 3, 4, 5])
sub_arr = arr[1:3].copy()
print(sub_arr)

sub_arr[0] = 200

print(sub_arr)
print(arr)
```

## Common Data Type
- Integers: int32, int64
- Floating point nums: float32, float64
- Bloolean: bool
- Complex nums: complex64, complex128
- String: S (byte-str) & U (unicode-str)
- Object: generic python objects - object

```python
# Data Type

arr = np.array([1, 2, 3, 4, 5])

print(arr, arr.dtype)
```

```python
# Data Type

arr = np.array([1, 2, 3.14, 4, 5])

print(arr, arr.dtype)
```

```python
# Data Type

arr = np.array([1, 2, 3, 4, 5, "Prime"])

print(arr, arr.dtype)
```

```python
# How Complex numbers work

arr1 = np.array([3 + 5j])
arr2 = np.array([2 + 3j])

print(arr1 + arr2)
print(arr1 - arr2)
```

```python
# Object

arr = np.array(["Prime", {1, 2, 3}, 3.14])

print(arr, arr.dtype)
```
