# <u>Supervesed Learning</u>
Supervised ML trains algorithms on labeled dataset, meaning each input has a known correct output, to learn patterns and predict outcomes for new data.

### Types of problems
1. **Regression -** output is continuous numerical values
2. **Classification -** output is a category or class

## <u>Scikit-learn</u>
**sklearn** is an open source ML library that supports supervised and unsupervised learnin.

### <u>**Algorithms**</u>
- Linear Regression
- Logistic Regression
- KNN
- Decision Trees
- Naive Bayes
- SVM etc.
