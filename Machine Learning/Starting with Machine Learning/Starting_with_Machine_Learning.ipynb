# <u>Machine Learning</u>
ML is a part of computer science where we teach computers to **Learn from data,** find patterns & make decision or predictions.

### <u>Machine Learning Types</u>
1. Supervised Learning
- Labelled data is raw data (such as images, text, or audio) that has been paired with a meaningful tag, label, or annotation that informs a machine learning  model about the desired outcomes
2. Unsupervised Learning
- Unlabelled data is raw data that does not have predefined labels or outputs. The machine learning model analyzes the data to find hidden patterns, similarities, or groups on its own.
3. Reinforcement Learning
- Reinforcement Learning is a type of machine learning where an agent learns by interacting with an environment. It receives rewards for correct actions and penalties for wrong actions, improving its decisions over time.
