```python
# Clean & structure the data
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5 }
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        # Clean ratings - data consistency
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # Handle missing vals
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None
            
        # Deduplication
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])    
        cleaned_data.append(user)

    return cleaned_data
```

```python
data = clean_data(data)
```

## <u>Assignment 2</u>

Getting meaningful insights from data
1. What's avg rating?
2. What % of users gave poor rating (<3)?

```python
# Get meaningful insights from data
def get_insights(data):

    # avg ratin
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])

    print(f"Avrage rating = {tot_rating/len(data)}")

    # percentage of users with poor rating
    poor_ratings = 0
    for user in data:
        if(float(user["rating"]) <3):
            poor_ratings += 1

    print(f"Percentage of user with poor rating = {poor_ratings/len(data) * 100}%")
```

```python
get_insights(data)
```

## <u>Assignment 3</u>
Build a product recommendation feature (Rule bassed, not ML)

1. If user rating >= 4, recommend same brand products.
2. If user rating <4, recommend different brand product.

```python
# Build recommendation Feature
def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user ["name"]

        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Sumsung"

        recommendations.append(curr_recomm)
    return recommendations
```

```python
get_recommendations(data)
```
