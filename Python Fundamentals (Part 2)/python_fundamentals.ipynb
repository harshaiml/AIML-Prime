### Conditional Statement

Match Case

```python
color = input("Enter color: ")

match color:
    case "Green":
        print("Go")
    case "Yellow":
        print("Look")
    case "Red":
        print("Stop")
    case _:
        print("Worng color!")
```

## Loops

*while loop*

```python
# Finite loop => 5x

i = 1 #iterator
while (i <= 5):
    print("Hello world", i)
    i += 1
print("After loop, count =", i) #11
```

```python
# Print numbers 1 to 5.

i = 1
while (i <= 5):
    print(i)
    i += 1
```

```python
# Print numbers 1 to 5 in reverse

i = 5
while (i >= 1):
    print(i)
    i -= 1
```
