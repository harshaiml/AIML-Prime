### Conditional Statement

Match Case

```python
color = input("Enter color: ")

match color:
    case "Green":
        print("Go")
    case "Yellow":
        print("Look")
    case "Red":
        print("Stop")
    case _:
        print("Worng color!")
```

## Loops

*while loop*

```python
# Finite loop => 5x

i = 1 #iterator
while (i <= 5):
    print("Hello world", i)
    i += 1
print("After loop, count =", i) #11
```

```python
# Print numbers 1 to 5.

i = 1
while (i <= 5):
    print(i)
    i += 1
```

```python
# Print numbers 1 to 5 in reverse

i = 5
while (i >= 1):
    print(i)
    i -= 1
```

Print multiplication table of any number 'n'

```python
# Multiplication table of any number

n = int(input("Enter your number: "))
i = 1
while (i <= 10):
    print (n * i)
    i += 1
```

```python
# Multiplication table of any number

n = int(input("Enter your number: "))
i = 0
while (i < 10): # 0 to 9
    print (n * (i+1))
    i += 1
```

break & continue

```python
# break

i = 1

while (i <= 10):
    if (i % 6 == 0):
        break
    print(i)
    i += 1

print("Outside loop now....")
```

```python
# continue

i = 1

while (i <= 10):
    if (i % 3 == 0):
        i += 1
        continue
    print(i)
    i += 1

print("outside loop....")
```

```python
# odd numbers

i = 1

while (i <= 10):
    if (i % 2 == 0):
        i += 1
        continue
    print(i)
    i += 1
```

```python
# odd numbers

i = 1

while (i <= 10):
    i += 1
    if (i % 2 == 0):
        continue
    print(i)
```

Print vowel count of a given string.


```python
word = "artificial intelligence"
count = 0

for ch in word:
    if(ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u'):
        count +=1
print("Answer = ",count)
```
