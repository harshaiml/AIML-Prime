**Q2.** Write a function that takes two integers a and b and prints all even numbers between them (inclusive).

```python
def even_num_check(a, b):
    for i in range(a, b +1):
        if i % 2 == 0:
            print(i)

even_num_check(3, 10)
```

**Q3.** Write a function that prints the digits of a numbers, n.

For eg: n = 312, there are 3 digits in it 3, 1 and 2 & we need to print them.

**[Hint-** The right most digit of a numbers N is N%10.

And to remove the right most digit from a numbers, we can do N = N / 10.]

```python
def print_digit(n):
    while n >0:
        digit = n % 10
        print(digit)
        n = n // 10
print_digit(312)
```

```python
def print_digit(n):
    if n == 0:
        return
    print_digit(n // 10)
    print(n % 10)
print_digit(512)
```

**Q4.** Write a function to return the count the number of digits in a number, n.

```python
def count_digit(n):
    count = 0
    while n> 0:
        n = n // 10
        count += 1
    return count

num = int(input("Enter a number: "))
print("You enter this number: ",num)
print("Total digit", count_digit(num))
```

**Q5.** Write a function to return the sum of digits of a number, n.

```python
# Using while loop.

def sum_of_digits(n):
    total = 0
    
    while n> 0:
        digit = n % 10         # Last digit
        total += digit         # Add in sum
        n = n // 10            # Remove last digit

    return total

# Function call
num = int(input("Enter your number: "))
print("You enterd this number",num)
print("Sum of digits",sum_of_digits(num))
```

```python
# Using for loop.

def sum_of_digits(n):
    total = 0

    for digit in str(n):
        total += int(digit)

    return total

num = int(input("Enter number: "))
print("You enterd this number",num)
print("Sum = ", sum_of_digits(num))
```
