## <b>Assignment Problem</b>

<b>Q1.</b> Start by loading the IRIS dataset. [<i>Dataset Link:</i> <u>Iris Flower Dataset</u>]

<b>A.</b> Display the following information:
- First 10 rows
- Shape & Data types
- Summary statistics (mean, std, min, max)

<b>B.</b> Select only those rows where:
- petal_length > 4.5
- species = "Iris-virginica"

<b>C.</b> Group by species and compute:
- Average sepal_lenght
- Maximum petal_width
- Standard deviation of sepal_width

<b>D.</b> Create a new column: <b>"petal_ratio" = petal_lenght / petal_width</b>
- For each species, find the average petal_ratio.

```python
import pandas as pd

df = pd.read_csv("IRIS.csv")

# Step A
print(df.head(10))
print(df.shape)
print(df.dtypes)
print(df.describe())

# Step B
filtered_df = df[(df['petal_length'] > 4.5) & (df['species'] == 'Iris-virginica')]
print(filtered_df)

# Step C
print(df.groupby('species')['sepal_length'].mean())
print(df.groupby('species')['petal_width'].max())
print(df.groupby('species')['sepal_width'].std())

# Step D
df['petal_ratio'] = df['petal_length'] / df['petal_width']
print(df.groupby('species')['petal_ratio'].mean())
```
