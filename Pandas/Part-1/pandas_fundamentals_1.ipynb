### Filtering Data

```python
# Filtering of data

df[ df["aqi"] > 100]
df[ (df["aqi"] > 100) & (df["temperature"] > 30)]
df[ (df["aqi"] > 100) & (df["temperature"] < 30)]

aqi_data = df[ (df["aqi"] > 100) & (df["temperature"] > 30) ][["city", "aqi"]]

aqi_data.iloc[0]
aqi_data.loc[6]
```

query() rules
- Expression as a string
- Column based reference
- Backticks for col_name with space or special chars
- Operators (&, |,~,>=,>,<=,<,== etc,)
- @ to referance Python vars
- Chained comparisons

```python
# Query

# df.query("aqi > 100 & temperature > 30")[["city","aqi"]]

aqi_val = 100
df.query("aqi > 100 & temperature > 30")[["city", "aqi"]]
aqi_data = df.query("aqi > @aqi_val & temperature > 30")[["city","aqi"]].copy()
```

Cleaning Data
1. Handle Missing Values
- isnull() / isna()
- isnull().sum()
- dropna()
- fillna()
- fill()
- bfill()

```python
# Cleaning Data - Missing values

df = pd.read_csv("raw_data.csv")

df.isnull()
df.isna()

df.isnull().sum()

df.dropna() # row drop
df.dropna(axis=1) # col drop

df.fillna(0)

age_mean = df["age"].mean()

cleaned_data = df.copy()
age_mean = cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)

df.ffill()
df.bfill()
```

### <u>Cleaning Data</u>
Handel Duplicate Values
- duplicated()
- drop_duplicates()

```python
# Handel Duplicate

df[["country", "gender"]].duplicated()

df.drop_duplicates(inplace=True)
```

### <u>Cleaning Data</u>

Handle Data Types & data-time

```python
# Data Types

df.dtypes
df2 = df.copy()

df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64").copy()
df2.dtypes

df["gender"].str.capitalize()

df["name"].str.split(" ")

df["country"].str.contains("US")
```
