### <u>Constructor</u>
_init_Method

```python
class Student:
    def __init__(self):
        print("constructor was called...")

stu1 = Student()
stu2 = Student()
stu3 = Student()
stu4 = Student()
```

```python
class Student:
    def __init__(self, name):
        self.name = name

stu1 = Student("Rahul")
stu2 = Student("Urvashi")
stu3 = Student("Shradha")
stu4 = Student("Harsh")

print(stu1.name)
print(stu2.name)
print(stu3.name)
print(stu4.name)
```

```python
class Student:
    def __init__(self, name, cgpa):
        self.name = name
        self.cgpa = cgpa

stu1 = Student("Rahul", 9.8)
stu2 = Student("Urvashi", 8.5)
stu3 = Student("Shradha", 7.9)
stu4 = Student("Harsh", 8.3)

print(stu1.name, stu1.cgpa)
print(stu2.name, stu2.cgpa)
print(stu3.name, stu3.cgpa)
print(stu4.name, stu4.cgpa)
```

```python
class Student:
    def __init__(self, name, cgpa):
        self.name = name
        self.cgpa = cgpa
    
    def get_cgpa(self):
        return self.cgpa
    
stu1 = Student("Rahul", 9.8)
stu2 = Student("Urvashi", 8.5)
stu3 = Student("Shradha", 7.9)
stu4 = Student("Harsh", 8.3)

print(f"{stu1.name} has cgpa = {stu1.get_cgpa()}")
```
