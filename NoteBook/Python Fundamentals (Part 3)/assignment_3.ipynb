**Q1.** Ask the user for string and check whether it is a palindrome or not.

A **palindrom** is a string which is same when we read it forward & backward. Eg - "madam", "rececar" etc.

**[Hint -** A palindrome string is equal to the reversed version of the string. We can use a loop to reverse the string manually.**]**

```python
# Take input from user
text = input("Enter a string: ")

# Reverse manually using loop
reverse_text = ""
for char in text:
    reverse_text = char + reverse_text

# Check palindrome
if text == reverse_text:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

**Q2.** Given a list of integers compute the average of all numbers in the list.

```python
# List of numbers
numbers = [10, 20, 30, 40]

# Step 1: Initialize sum
total = 0

# Step 2: Loop to add all  numbers
for num in numbers:
    total += num

# Step 3: Calculate average
average = total / len(numbers)

# Step 4: Print result
print("Avrage is =",average)
```

**Q3.** Input two lists of integers from the user. Marge them into one list and sort the result.

Eg - list1 = [1, 2, 7], list2 = [2, 4, 5]

result = [1, 2, 3, 4, 5, 7]

```python
# Step 1: Take input from user
list1 = list(map(int, input("Enter first list elements").split()))
list2 = list(map(int, input("Enter second list elements:").split()))

# Step 2: Marge both lists
result = list1 + list2

# Step 3: Sort the merged list
result.sort()

# Step 4: Print final result:
print("Merged & Sorted List",result)
```

**Q4.** Given a tuple of integers, create:

- A tuple of all even numbers
- A tuple of all odd numbers

```python
# Given tuple
t =(1, 2, 3, 4, 5, 6)

even = []
odd = []

for num in t:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

even_tuple = tuple(even)
odd_tuple = tuple(odd)

# Output
print("Even numbers:", even_tuple)
print("Odd numbers:",odd_tuple)
```

**Q5.** Create a dictionary where:
- Keys = student names
- Values = marks (integer)

Write a menu-based program where user presses a key ('A','B','C','D')
depending on the operation they want to perform on the dictionary:
1. A - Add a student
2. B - Update marks
3. C - Search for a student
4. D - Display all students and marks

```python
students = {}

while True:
    print("\nMenu:")
    print("A - Add a student")
    print("B - Update marks")
    print("C - Search student")
    print("D - Display all")
    print("E - Exit")

    choice = input("Enter your choice: ").upper()

    # A → Add student
    if choice == 'A':
        name = input("Enter student name: ")
        marks = int(input("Enter marks: "))
        students[name] = marks
        print("Student added successfully!")

    # B → Update marks
    elif choice == 'B':
        name = input("Enter student name: ")
        if name in students:
            marks = int(input("Enter new marks: "))
            students[name] = marks
            print("Marks updated!")
        else:
            print("Student not found!")

    # C → Search student
    elif choice == 'C':
        name = input("Enter student name: ")
        if name in students:
            print(f"{name}'s marks: {students[name]}")
        else:
            print("Student not found!")

    # D → Display all
    elif choice == 'D':
        if len(students) == 0:
            print("No data available")
        else:
            for name, marks in students.items():
                print(name, ":", marks)

    # E → Exit
    elif choice == 'E':
        print("Program exited")
        break

    else:
        print("Invalid choice, try again!")
```
