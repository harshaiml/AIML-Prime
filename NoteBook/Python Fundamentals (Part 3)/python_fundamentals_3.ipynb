### <u>Lists</u>

Methods

- l.append(val)            # Add one element at the end
- l.insert(idx, val)       # Insert element at idx
- l.sort()                 # Arranges in increasing order
- l.reverse()              # Reverse order

```python
# l.append(val)            # Add one element at the end

nums = [1, 2, 3]
nums.append(4)
print(nums)
```

```python
# l.insert(idx, val)       # Insert element at idx

nums = [ 1, 2, 3]
nums.insert(1,8)
print(nums)
```

```python
# l.sort()                 # Arranges in increasing order

nums = [ 1, 2, 3]
nums.insert(1,8)
print(nums)

nums.sort()
print(nums)
```

```python
# l.sort(reverse=True)         # Arranges in increasing order in reverse

nums = [ 1, 2, 3]
nums.insert(1,8)
print(nums)

nums.sort(reverse=True)
print(nums)
```

```python
# l.reverse()              # Reverse order

nums = [1, 2, 3, 4, 5]
nums.reverse()
print(nums)
```

### <u>Loops with list </u>

```python
nums = [1, 2, 3, 10, 4]

for val in nums:
    print(val)
```

```python
nums = [1, 2, 3, 10, 4]

x = int(input("Enter your value: "))
idx = 0

for val in nums:
    if (val == x):
        print(f"{x} found at index= {idx}")
        break
    idx +=1
else:
    print("Not found")
```
