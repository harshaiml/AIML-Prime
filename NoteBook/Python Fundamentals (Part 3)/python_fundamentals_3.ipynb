### Indexing

```python
word = "Python"

print(word[2])
```

```python
word = "Python"
for ch in word:
    print(ch)
```

### Slicing

```python
word = "Python"
print(word[2:4])
```

```python
word = "I am learning Python"
print(word[14:20])
```

```python
word = "I am learning Python"
print(word[14:])
```

```python
word = "I am learning Python"
print(word[14:len(word)])
```

```python
word = "I am learning Python"
print(word[:len(word)])
```

```python
word = "I am learning Python"
print(word[:])
```

```python
# Reverse indexing

word = "Python"
print(word[-4:-2])
```

### Formatting

```python
a = 5
b = 10
sum = a + b

# Noramal formatting
print("Sum is {}".format(sum))
```

```python
a = 5
b = 10
sum = a + b

# Noramal formatting
print("Language is {}".format("Python"))
print("Sum of {} & {} is {}".format(a, b, sum))
```

```python
# Index based formattinh
a = 5
b = 10
sum = a + b

print("Sum of {1} & {0} is {2}".format(a, b, sum))
```

```python
# Value based formatting

print("Value of vars {a} & {b}".format(a=5, b=10)) 
```
