### <u>Lists</u>

Methods

- l.append(val)            # Add one element at the end
- l.insert(idx, val)       # Insert element at idx
- l.sort()                 # Arranges in increasing order
- l.reverse()              # Reverse order

```python
# l.append(val)            # Add one element at the end

nums = [1, 2, 3]
nums.append(4)
print(nums)
```

```python
# l.insert(idx, val)       # Insert element at idx

nums = [ 1, 2, 3]
nums.insert(1,8)
print(nums)
```

```python
# l.sort()                 # Arranges in increasing order

nums = [ 1, 2, 3]
nums.insert(1,8)
print(nums)

nums.sort()
print(nums)
```

```python
# l.sort(reverse=True)         # Arranges in increasing order in reverse

nums = [ 1, 2, 3]
nums.insert(1,8)
print(nums)

nums.sort(reverse=True)
print(nums)
```

```python
# l.reverse()              # Reverse order

nums = [1, 2, 3, 4, 5]
nums.reverse()
print(nums)
```

### <u>Dictionary</u>
Method

```python
# Dictionary

# Method

# d.keys()              # return all keys
# d.values()            # return all values
# d.items()             # return (key, val) pairs
# d.get(val)            # return val acc. to key
# d.update(new)         # adds new item to dict
```

```python
# d.keys()        # return all keys

info = {
    "name": "Harsh",
    "cgpa": 9.2,
    "subjects": ["Math", "Science"],
    3.14: "PI"
    
}
print(info.keys())
```

```python
# d.keys()        # return all keys

info = {
    "name": "Harsh",
    "cgpa": 9.2,
    "subjects": ["Math", "Science"],
    3.14: "PI"
    
}

dict_keys = info.keys()

print(dict_keys)
```

```python
# d.keys()        # return all keys

info = {
    "name": "Harsh",
    "cgpa": 9.2,
    "subjects": ["Math", "Science"],
    3.14: "PI"
    
}

dict_keys = list(info.keys())

print(type(dict_keys))
```

```python
# d.values()      # return all values


info = {
    "name": "Harsh",
    "cgpa": 9.2,
    "subjects": ["Math", "Science"],
    3.14: "PI"
    
}

dict_vals = list(info.values())
print(dict_vals)
```

```python
# d.items()       # return (key, val) pairs

info = {
    "name": "Harsh",
    "cgpa": 9.2,
    "subjects": ["Math", "Science"],
    3.14: "PI"
    
}

dict_items = list(info.items())
print(dict_items)
```

```python
# d.get(val)      # return val acc. to key

info = {
    "name": "Harsh",
    "cgpa": 9.2,
    "subjects": ["Math", "Science"],
    3.14: "PI"
    
}

print(info.get("cgpa"))   # worng key

print("End of code")
```

```python
# d.update(new)   # adds new item to dict

info = {
    "name": "Harsh",
    "cgpa": 9.2,
    "subjects": ["Math", "Science"],
    3.14: "PI"
    
}

info.update({
    "city": "Delhi"
})

print(info)
```

## <u>Sets</u>

Methods

```python
# s.add(val)               # adds a val
# s.remove(val)            # remove a val
# s.clear()                # empties the set
# s.pop()                  # removes a random val
# s.union(set2)            # returns new union
# s.intersection(set2)     # returns new intersection
```

```python
# s.add(val)               # adds a val

my_set = {1, 2, 2, 2, 3}

my_set.add(5)
print(my_set)
```

```python
# s.remove(val)            # remove a val

my_set = {1, 2, 2, 2, 3}

my_set.add(5)
my_set.remove(1)
print(my_set)
```

```python
# s.clear()                # empties the set

my_set = {1, 2, 2, 2, 3}

my_set.add(5)
my_set.remove(1)
my_set.clear()
print(my_set)
```

```python
# s.pop()                  # removes a random val

my_set = {1, 2, 2, 2, 3}
print(my_set)
my_set.pop()
print(my_set)
```

```python
# s.union(set2)            # returns new union

s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 8, 9, 10}

print(s1.union(s2))
```

```python
# s.intersection(set2)     # returns new intersection

s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 8, 9, 10}

print(s1.intersection(s2))
```

Given a list of tuples with info(name,subject):

```python
# list student enrolled in English

info = [
    ("Alice", "Math"),
    ("Bob", "Science"),
    ("Alice", "Science"),
    ("Charlie", "Math"),
    ("Bob", "Math"),
    ("Alice", "English"),
    ("Charlie", "English"),
]

for name,cource in info:
    if(cource == "English"):
        print(name)
```

```python
# create dictionary (student, set of courses)

info = [
    ("Alice", "Math"),
    ("Bob", "Science"),
    ("Alice", "Science"),
    ("Charlie", "Math"),
    ("Bob", "Math"),
    ("Alice", "English"),
    ("Charlie", "English"),
]

dict = {}

for name,cource in info:
    if(dict.get(name) == None):
        dict.update({name: set()})
        dict[name].add(cource)
    else:
        dict[name].add(cource)

print(dict)
```
