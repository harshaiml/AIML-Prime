### Exploring Methods with Movies Data
- title()
- xlabel()
- ylabel()
- legend()

```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The king's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

# Plot
plt.plot(years,oscar_revenue)

plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.show()
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avtar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

plt.plot(years, oscar_revenue, label="Oscar Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
plt.show()
```

### Format Strings

A format string consists of a part for color, marker and line:
           
fmt = '[marker][line][color]'
- color
- marker
- linestyle
- line width

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avtar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

plt.plot(years, oscar_revenue, "o--r", label="Oscar Movies")
plt.plot(years, non_oscar_revenue, "<-y", label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
plt.show()
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avtar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

plt.plot(years, oscar_revenue, color="#CBC3E3", marker="o", linestyle="--", linewidth=4, label="Oscar Movies")
plt.plot(years, non_oscar_revenue, color="pink", marker="<", linestyle="-.", linewidth=2, label="Non-Oscar Movies")

plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
plt.show()
```
