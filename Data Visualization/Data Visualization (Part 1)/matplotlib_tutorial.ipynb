### <u>Plot Methods</u>
1. plt.plot(x,y) - Line plot
2. plt.bar(x, height) - Vertical bar chart
3. plt.barh(y, width) - Horizontal bar chart
4. plt.scatter(x,y) - Scatter plot
5. plt.pie(data) - Pie chart
6. plt.hist(data, bins) - Histogram
7. plt.boxplot(data) - Box plot

### <u>Bar Chart</u>
A bar plot (or bar chart) is a graph that displays categorical using rectangular bars.

Each bar's height (or lenght, in horizontal bars) represents a value.

```python
# Bar Chart

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The king's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue, color="Aqua")
plt.title("Revenue in each Years for Oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# plt.text()

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The king's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue, color="Aqua")
plt.title("Revenue in each Years for Oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.text(years[0], oscar_revenue[0]+20, str(oscar_revenue[0]), ha="center")
```

```python
# plt.text() use with loop for all bars.

movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The king's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue, color="Aqua")
plt.title("Revenue in each Years for Oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
for  i in range(len(years)):
    plt.text(years[i], oscar_revenue[i]+20, str(oscar_revenue[i]), ha="center")

plt.ylim(0, max(oscar_revenue)+75)
```

### Compare 2 sets of Data

We customise width. Width controls how thick each bar is. Usually ranges from 0 to 1.

[Default val is 0.8]

```python
import numpy as np

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The king's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avtar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M

years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years))
width = 0.35

plt.bar(x-width/2, oscar_revenue, width, label="Oscar Movies") # Shifting to LEFT
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar Movies") # Shifting to RIGHT

plt.title("Oscar vs Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue ($M)")
plt.legend()

plt.xticks(x, years)
plt.show()
```

## <u>Scatter Plots</u>
Annotations

plt.annotate("text",xy=(_point, y_point), xytext=(x_text, y_text))

```python
# Scatter Plots

people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]

age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if x < 130 else "red" for x in bp]

plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
plt.title("Age VS BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
# plt.grid()
plt.colorbar(label="BP")

for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], bp[i]), xytext=(age[i] +1, bp[i]+1))

plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+5)
plt.show()
```

## <u>Pie Charts</u>
A Pie Charts is a circualr chart divided into slices to show relative proportions of a whole.
- Each slice = a category
- Size of slice = valueof that category relative to total

Prefer them only for <b>few categories</b>

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

plt.style.use("dark_background")
plt.pie(amounts, labels=expenses)
```

## <u>Pie Charts</u>
Costomizations
- <b>autopct="%1.1%%"</b> - shows percentage of each slice
- <b>colors</b> - add custom colours
- <b>Wedgeprops</b> - define (edgecolor, linewidth, linestyle, alpha, fill...)
- <b>explod</b> - highlights a slice by pulling it out
- <b>startangle</b> - rotate the chart
- <b>shadow</b> - adds 3D effect or makes the chat circular
