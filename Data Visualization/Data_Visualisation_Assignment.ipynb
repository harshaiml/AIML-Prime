### Q1. For the following dataset generated using NumPy:

np.random.normal(loc=70, scale=10, size=200)
- Plot a histogram of scores
- Draw a vertical line for the mean & try to interpret the results.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random dataset
data = np.random.normal(loc=70, scale=10, size=200)

# Step 2: Calculate mean
mean = np.mean(data)

# Step 3: Create histogram
plt.hist(data, bins=15, color="skyblue", edgecolor="black")

# Step 4: Draw vertical line for mean
plt.axvline(mean, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean:.2f}")

# Step 5: Add title and labels
plt.title("Histogram of Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

# Step 6: Show legend
plt.legend()

# Step 7: Display graph
plt.show()
```

## Interpretation

- Most scores are concentrated around the mean value.
- The distribution appears approximately normal.
- Very high and very low scores are less frequent.
