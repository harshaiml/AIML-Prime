### Q1. For the following dataset generated using NumPy:

np.random.normal(loc=70, scale=10, size=200)
- Plot a histogram of scores
- Draw a vertical line for the mean & try to interpret the results.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random dataset
data = np.random.normal(loc=70, scale=10, size=200)

# Step 2: Calculate mean
mean = np.mean(data)

# Step 3: Create histogram
plt.hist(data, bins=15, color="skyblue", edgecolor="black")

# Step 4: Draw vertical line for mean
plt.axvline(mean, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean:.2f}")

# Step 5: Add title and labels
plt.title("Histogram of Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

# Step 6: Show legend
plt.legend()

# Step 7: Display graph
plt.show()
```

## Interpretation

- Most scores are concentrated around the mean value.
- The distribution appears approximately normal.
- Very high and very low scores are less frequent.

### Q2. For the Seaborn penguins dataset, do the following:
- Plot flipper_length_mm vs body_mass_g
- Color by species

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
penguins = sns.load_dataset("penguins")

# Step 2: Create scatter plot
sns.scatterplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)

# Step 3: Add title and labels
plt.title("Flipper Length vs Body Mass")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")

# Step 4: Show plot
plt.show()
```

## Interpretation

- Penguins with larger flipper lengths generally have higher body mass.
- Different species are represented using different colors.
- Gentoo penguins appear larger compared to other species.

### Q3. For the Seaborn tips dataset, do the following:
- Create a box plot of total_bill for each day
- Add hue="sex"
- Identify which day has the highest median bill
- Comment on outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
tips = sns.load_dataset("tips")

# Step 2: Create box plot
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

# Step 3: Add title and labels
plt.title("Total Bill Distribution by Day and Sex")
plt.xlabel("Day")
plt.ylabel("Total Bill")

# Step 4: Show plot
plt.show()
```

## Interpretation

- Saturday appears to have the highest median total bill.
- Male customers generally have slightly higher bill distributions.
- Some outliers are visible, especially on weekends, indicating unusually large bills.

### Q4. For the Seaborn tips dataset, do the following:
- Create a scatter plot of total_bill vs tip
- Color points by sex
- Use different markers for smoker
- Add a title and axis labels
- Analyze: Do smokers tend to tip differently?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Create scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)

# Title and labels
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

## Interpretation

- Higher total bills generally lead to higher tips.
- Male and female customers both follow a similar tipping pattern.
- Smokers and non-smokers do not show a huge difference in tipping behavior.
- Some customers gave unusually high tips for large bills.
