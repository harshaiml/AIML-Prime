**Q3.** Ask the user to enter two integers and one float. Convert them all to floats and print their average.

```python
# Q3. Ask the user to enter two integers and one float. Convert them all to floats and print their average.

# Taking input
num1 = int(input("Enter your first integer number:"))
num2 = int(input("Enter your second integer number:"))
num3 = float(input("Enter your float number:"))

# Converting all to float
num1 = float(num1)
num2 = float(num2)

# Calculating average
avg = (num1 + num2 + num3)/3

# Printing result
print("Your first integer number is", num1)
print("Your second integer number is", num2)
print("Your float number is", num3)
print("Average is", avg)
```

**Q6.** Write a program to swap values of two numbers entered by the user.

```python
# **Q6** Write a program to swap values of two numbers entered by the user.

num1 = int(input("Enter your first number:"))
num2 = int(input("Enter yurr second number:"))

# Step 1: store num1 in temp
temp = num1

# Step 2: assign num2 to num1
num1 = num2

# Step 3: assign temp (old num1) to num2
num2 = temp

print("After swapping:")
print("num1",num1)
print("num2",num2)
```

**Q7.** Ask the user for a temperature in Celsius (string input). Convert it to float, then calculate and print temperature in Fahrenheit.

Conversion formula:  *FahrennheitTemp = (CelsiusTemp * (9/5)) + 32*

```python
# Take input (string)
celsius = input("Enter temperature in Celsius:")

# Convert to float
celsius = float(celsius)

# Apply formula
fahrenheit = (celsius*(9/5))+32

# Print result
print("Your temperature is",celsius)
print("Temperature in Fahrenheit is",fahrenheit)
```
