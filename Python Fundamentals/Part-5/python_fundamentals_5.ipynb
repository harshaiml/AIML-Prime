## <u>File Operations</u>
- r ------------> reading [default]
- w ------------> writing, truncates file first
- x ------------> creates new & open for writing
- a ------------> writing. appends at end
- b ------------> binary mode
- t ------------> text mode [default]
- #+------------> opens disk file for update (r & w)

```python
## <u>File Operations</u>
# r    # reading [default]
# w    # writing, truncates file first
# x    # creates new & open for writing
# a    # writing. appends at end
# b    # binary mode
# t    # text mode [default]
# +    # opens disk file for update (r & w)
```

## <u>Word Search</u>

```python
data = True

with open("sample.txt", "r") as f:
    while data:
        data = f.readline()
        if("Python" in data):
            print("word found")
            break
        print(data)
```
