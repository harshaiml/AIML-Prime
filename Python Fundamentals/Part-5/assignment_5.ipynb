**Q1.** Create a program that:
1. Opens a file "name.txt" in write mode
2. Writes 5 names (one per line) entered by the user
3. Then opens the same filein read mode and prints all names

```python
# Step 1 & 2: Write 5 names into file
with open("name.txt","w") as file:
    for i in range(5):
        name = input(f"Enter name {i+1}:")
        file.write(name + "\n")

# Step 3: Read and print all names
with open("name.txt","r") as file:
    data = file.read()
    print("\nNames in file:")
    print(data)
```

**Q2.** Create a program that:
1. Opens a file *"log.txt"* in append mode
2. Adds a new log entry (like "Program run successfully")
3. Opens the file in read mode and prints all logs

```python
# Step 1: Open file in append mode and add log
with open("log.txt","a") as f:
    f.write("Program run successfully\n")

# Step 2: Open same file in read mode and print all logs
with open("log.txt", "r") as f:
    data = f.read()
    print(data)
```

**Q3.** Create a program that:
1. Has a list of numbers: [5, 10, 15, 20, 25]
2. Uses a list comprehension to create a new list with only numbers greater than 15
3. Print the new list

```python
# Step 1:  Given list
nums = [5, 10, 15, 20, 25]

# Step 2: List comprehension (filter numbers > 15)
new_list = [num for num in nums if num > 15]

# Step 3: Print result
print(new_list)
```
