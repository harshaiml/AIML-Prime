**Q1.** Create a program that:
1. Opens a file "name.txt" in write mode
2. Writes 5 names (one per line) entered by the user
3. Then opens the same filein read mode and prints all names

```python
# Step 1 & 2: Write 5 names into file
with open("name.txt","w") as file:
    for i in range(5):
        name = input(f"Enter name {i+1}:")
        file.write(name + "\n")

# Step 3: Read and print all names
with open("name.txt","r") as file:
    data = file.read()
    print("\nNames in file:")
    print(data)
```

**Q2.** Create a program that:
1. Opens a file *"log.txt"* in append mode
2. Adds a new log entry (like "Program run successfully")
3. Opens the file in read mode and prints all logs

```python
# Step 1: Open file in append mode and add log
with open("log.txt","a") as f:
    f.write("Program run successfully\n")

# Step 2: Open same file in read mode and print all logs
with open("log.txt", "r") as f:
    data = f.read()
    print(data)
```

**Q3.** Create a program that:
1. Has a list of numbers: [5, 10, 15, 20, 25]
2. Uses a list comprehension to create a new list with only numbers greater than 15
3. Print the new list

```python
# Step 1:  Given list
nums = [5, 10, 15, 20, 25]

# Step 2: List comprehension (filter numbers > 15)
new_list = [num for num in nums if num > 15]

# Step 3: Print result
print(new_list)
```

**Q4.** Create a Python dictionary of 3 cities and their populations. Save it to "cities.json".
1. Then load the JSON and print each city and its population.
2. ASk the user for a new city & its population - update this info in the json file.

```python
import json

# Step 1: Create dictionary
cities = {
    "Delhi": 20000000,
    "Mumbai": 18000000,
    "Lucknow": 3000000,
}

# Step 2: Save to JSON file
with open("cities.json", "w") as f:
    json.dump(cities, f)

# Step 3: Load JSON  and print data
with open("cities.json", "r") as f:
    data = json.load(f)
    print("Cities and Population:")
    for city, population in data.items():
        print(city, ":", population)

# Step 4: Take user input
new_city = input("Enter new city: ")
new_population = int(input("Enter population: "))

# Step 5: Update data
data[new_city] = new_population

# Step 6: Save updated data back to JSON
with open("cities.json", "w") as f:
    json.dump(data, f)

print("Data updated successfully!")
```

**Q5.** Write a program that tries to open "data.txt" in read mode. if the file does not exite, catch the exception and print "File not found!".

```python
try:
    f = open("data.txt", "r")
    data = f.read()
    print(data)
    f.close()

except FileNotFoundError:
    print("File not found!")
```
