**Q1.** Create a program that:
1. Opens a file "name.txt" in write mode
2. Writes 5 names (one per line) entered by the user
3. Then opens the same filein read mode and prints all names

```python
# Step 1 & 2: Write 5 names into file
with open("name.txt","w") as file:
    for i in range(5):
        name = input(f"Enter name {i+1}:")
        file.write(name + "\n")

# Step 3: Read and print all names
with open("name.txt","r") as file:
    data = file.read()
    print("\nNames in file:")
    print(data)
```
