**Q1.** Ask the user for string and check whether it is a palindrome or not.

A **palindrom** is a string which is same when we read it forward & backward. Eg - "madam", "rececar" etc.

**[Hint -** A palindrome string is equal to the reversed version of the string. We can use a loop to reverse the string manually.**]**

```python
# Take input from user
text = input("Enter a string: ")

# Reverse manually using loop
reverse_text = ""
for char in text:
    reverse_text = char + reverse_text

# Check palindrome
if text == reverse_text:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

**Q2.** Given a list of integers compute the average of all numbers in the list.

```python
# List of numbers
numbers = [10, 20, 30, 40]

# Step 1: Initialize sum
total = 0

# Step 2: Loop to add all  numbers
for num in numbers:
    total += num

# Step 3: Calculate average
average = total / len(numbers)

# Step 4: Print result
print("Avrage is =",average)
```

**Q4.** Given a tuple of integers, create:

- A tuple of all even numbers
- A tuple of all odd numbers

```python
# Given tuple
t =(1, 2, 3, 4, 5, 6)

even = []
odd = []

for num in t:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

even_tuple = tuple(even)
odd_tuple = tuple(odd)

# Output
print("Even numbers:", even_tuple)
print("Odd numbers:",odd_tuple)
```

**Q5.** Create a dictionary where:
- Keys = student names
- Values = marks (integer)

Write a menu-based program where user presses a key ('A','B','C','D')
depending on the operation they want to perform on the dictionary:
1. A - Add a student
2. B - Update marks
3. C - Search for a student
4. D - Display all students and marks

```python
students = {}

while True:
    print("\nMenu:")
    print("A - Add a student")
    print("B - Update marks")
    print("C - Search student")
    print("D - Display all")
    print("E - Exit")

    choice = input("Enter your choice: ").upper()

    # A → Add student
    if choice == 'A':
        name = input("Enter student name: ")
        marks = int(input("Enter marks: "))
        students[name] = marks
        print("Student added successfully!")

    # B → Update marks
    elif choice == 'B':
        name = input("Enter student name: ")
        if name in students:
            marks = int(input("Enter new marks: "))
            students[name] = marks
            print("Marks updated!")
        else:
            print("Student not found!")

    # C → Search student
    elif choice == 'C':
        name = input("Enter student name: ")
        if name in students:
            print(f"{name}'s marks: {students[name]}")
        else:
            print("Student not found!")

    # D → Display all
    elif choice == 'D':
        if len(students) == 0:
            print("No data available")
        else:
            for name, marks in students.items():
                print(name, ":", marks)

    # E → Exit
    elif choice == 'E':
        print("Program exited")
        break

    else:
        print("Invalid choice, try again!")
```

**Q6.** Given a list of words:

words = ["apple", "banan", "kiwi", "cherry", "mango"]

Create a dictionary that maps each word to its lenght.

{"apple": 5, "banana: 6, "kiwi": 4, ....}

```python
words = ["apple", "banana", "kiwi", "cherry", "mango"]

result = {}

for word in words:
    result[word] = len(word)

print(result)
```

**Q7.** Write a program that takes a string from the user and prints the number of spaces in the string.

```python
# Step 1: Input from user
text = input("Enter a string: ")

# Step 2: Create counter
count = 0

# Step 3: Loop run
for ch in text:
    if ch == " ":
        count += 1

# Step 4: Output
print(text)
print("number of spaces:",count)
```

**Q8.** Write a program to check whether two list share no common elements.
- Share no common elements list1 = [1, 2, 3, 4] list2 = [5, 6, 7, 8]
- Share common elements list1 = [1, 2, 3] list2 = [3, 4]

[Hint - use sets]

```python
# Input lists
list1 = [1, 2, 3, 4]
list2 = [5, 6, 2, 8]

# Step 1: Convert lists to sets
set1 = set(list1)
set2 = set(list2)

# Step 2: Find common elements (intersection)
common = set1 & set2

# Step 3: Check result
if len(common) == 0:
    print("No common elements")
else:
    print("Common elements exist",common)
```

**Q9.** Given a list, print all elements that appear more than once in the list.

[Hint - use sets]

```python
# Input list
nums = [1, 2, 3, 2, 4, 5, 1]

seen = set()
duplicates = set()

for num in nums:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)

# Output
print("Seen elements",seen) # If We will remove duplicates values.
print("Duplicate elements:", duplicates)
```
